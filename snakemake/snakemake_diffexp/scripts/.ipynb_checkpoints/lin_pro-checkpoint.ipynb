```python
expressionData = pd.read_csv("cfDNA-reanalysis_Graz/kristin/cfDNA/resources/protein_atlas/RNAtableExtended.tsv.gz", index_col = 0 ,compression = "gzip", header = 0, sep = "\t")
```

```python
pd.set_option('display.max_columns', None)
```

```python
union = expressionData.index.intersection(pred_IH01.index)
expressionData = expressionData.loc[union]
expressionData
```

# Cell samples

```python
tissues = ["monocytes", "lung", "colon", "liver"]
```

```python
expressionTissues = expressionData.loc[:,tissues]
expressionTissues
```

```python
expressionTissues = np.log2(expressionTissues + 1).fillna(0)
```

```python
expressionTissues
```

```python
pd.DataFrame(pred_IH01)
```

```python
##need more columns, less rows 
##
```

```python
params, residuals = nnls(A = np.array(expressionTissues), b = np.array(pred_IH01), maxiter = 1000)
```

```python
for i, p in enumerate(params):
    print(expressionTissues.columns[i], ": ", p)
```

```python
res = lsq_linear(A =  np.array(expressionTissues), b = np.array(pred_IH01), bounds = (0,np.inf))
```

```python
for i, p in enumerate(res.x):
    print(expressionTissues.columns[i], ": ", p)
```
